```python
import os
import glob

def borrar_pngs(directorio):
    # Construir la ruta para los archivos .png
    archivos_png = glob.glob(os.path.join(directorio, "*.png"))
    
    # Borrar cada archivo .png encontrado
    for archivo in archivos_png:
        try:
            os.remove(archivo)
            print(f"Archivo borrado: {archivo}")
        except Exception as e:
            print(f"Error al borrar el archivo {archivo}: {e}")

# Especifica el directorio aqu
directorio = "F:\Github\Improving-deep-neural-networks-to-identify-mental-disorders-using-Neural-Architecture-Search\D3T3C\D3TEC Dataset\iPhoneSE2020"

borrar_pngs(directorio)

# Especifica el directorio aqu
directorio = "F:\Github\Improving-deep-neural-networks-to-identify-mental-disorders-using-Neural-Architecture-Search\D3T3C\D3TEC Dataset\SM-27"

borrar_pngs(directorio)
```
