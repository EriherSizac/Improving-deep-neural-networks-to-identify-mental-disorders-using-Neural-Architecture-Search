```python
import os
import glob

root_path = "d:\\Sistema\\Escritorio\\Escritorio\\Tesis\\DAIC-WOZ\\data"
for folder_name in os.listdir(root_path):
    folder_path = os.path.join(root_path, folder_name)
    if os.path.isdir(folder_path):
        for preprocessed_file in glob.glob(os.path.join(folder_path, '*.wav')):
            # Determine the target directories
            target_dir_resized = os.path.join(folder_path, 'resized')
            target_dir_normal = os.path.join(folder_path, 'normal')
            
            # Function to delete all files in a directory
            def delete_files_in_directory(directory):
                if os.path.exists(directory):
                    for file_name in os.listdir(directory):
                        file_path = os.path.join(directory, file_name)
                        try:
                            if os.path.isfile(file_path) or os.path.islink(file_path):
                                os.remove(file_path)
                                print(f'Deleted file: {file_path}')
                            elif os.path.isdir(file_path):
                                shutil.rmtree(file_path)
                                print(f'Deleted directory and its contents: {file_path}')
                        except Exception as e:
                            print(f'Failed to delete {file_path}. Reason: {e}')
            
            # Delete files in the target directories
            delete_files_in_directory(target_dir_resized)
            delete_files_in_directory(target_dir_normal)
            
            # Check if the target directory exists, if not, create it
            if not os.path.exists(target_dir_resized):
                os.makedirs(target_dir_resized)
                print(f'Created directory: {target_dir_resized}')
            
            if not os.path.exists(target_dir_normal):
                os.makedirs(target_dir_normal)
                print(f'Created directory: {target_dir_normal}')
```
