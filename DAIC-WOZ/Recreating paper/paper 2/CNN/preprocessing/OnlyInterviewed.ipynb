```python
# Global imports
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
from PIL import Image
import re
from mpl_toolkits.mplot3d import Axes3D
import librosa
import librosa.display
import pandas as pd
from pydub import AudioSegment
```

```python
# Function to save progress
def save_progress( wav_file_path):
    with open(PROGRESS_PATH, 'w') as file:
        file.write(wav_file_path)

# Function to load progress
def load_progress():
    progress_file = PROGRESS_PATH
    if os.path.exists(progress_file):
        with open(progress_file, 'r') as file:
            return file.read().strip()
    else:
        return None
```

```python
def extract_numbers_from_string(input_string):
    # Regular expression to match numbers
    pattern = r'\d+'
    # Find all numbers in the input string
    numbers = re.findall(pattern, input_string)
    return numbers
```

```python
def clean_audio(csv_file, wav_file, output_file):
    # Read the CSV file with tab-separated values
    transcript = pd.read_csv(csv_file, delimiter='\t')

    # Load the audio file
    audio = AudioSegment.from_wav(wav_file)

    # Initialize an empty AudioSegment for the cleaned audio
    cleaned_audio = AudioSegment.empty()

    # Iterate through the rows of the transcript
    for index, row in transcript.iterrows():
        # Check if the speaker is the participant
        if row['speaker'] == 'Participant':
            # Extract the start and stop times
            start_time = row['start_time'] * 1000  # convert to milliseconds
            stop_time = row['stop_time'] * 1000    # convert to milliseconds

            # Extract the segment and append to the cleaned audio
            cleaned_audio += audio[start_time:stop_time]

    # Export the cleaned audio
    cleaned_audio.export(output_file, format="wav")

    print(f"Cleaned audio saved as {output_file}")

```

```python
# Path to the directory containing folders with .wav files
root_path = "d:\Sistema\Escritorio\Escritorio\Tesis\DAIC-WOZ\data"

# Define the root path for saving progress
PROGRESS_PATH = os.path.join(os.getcwd(), 'progress.txt')
```

```python

# Load progress
last_wav_file_path = load_progress()
# Seek and get
for folder_name in os.listdir(root_path):
    folder_path = os.path.join(root_path, folder_name)
    number_file = extract_numbers_from_string(folder_name)[0]
    
    if os.path.isdir(folder_path):
        print(folder_path)
        print(last_wav_file_path)
        csv_file = f'{folder_path}/{number_file}_TRANSCRIPT.csv'
        tor = csv_file
        # Cut those without noise
        wav_file_unnoised = f'{folder_path}/preprocessed_{number_file}_AUDIO.wav'
        output_file_unnoised = f'{folder_path}/cleaned_preprocessed_{number_file}_AUDIO.wav'
        clean_audio(csv_file, wav_file_unnoised, output_file_unnoised)
        # Now, same for original audios
        wav_file_noised = f'{folder_path}/{number_file}_AUDIO.wav'
        output_file_noised = f'{folder_path}/cleaned_{number_file}_AUDIO.wav'
        clean_audio(csv_file, wav_file_noised, output_file_noised)
        print(f'Finished with {folder_name}')
        
```
